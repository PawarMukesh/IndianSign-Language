### Build CNN Architecture

```python
# Object creation
cnn_model = Sequential()

# Conv1: Add convolutional layer with relu activation with same padding and define input image size
cnn_model.add(Conv2D(32, kernel_size=(3,3),padding='same',activation='relu',input_shape=(224,224,3)))
cnn_model.add(MaxPooling2D(pool_size=(2,2))) # Add maxpooling layer to extract maximun feature

# Conv2: Add 2nd convolutional layer with relu activation with same padding 
cnn_model.add(Conv2D(64,kernel_size=(3,3),padding='same',activation='relu'))
cnn_model.add(MaxPooling2D(pool_size=(2,2)))

# Conv3: Add 3rd Convolutional layer with relu activation function with same padding
cnn_model.add(Conv2D(128, kernel_size=(3,3), padding='same',activation='relu'))
cnn_model.add(MaxPooling2D(pool_size=(2,2)))

cnn_model.add(BatchNormalization())

# Conv4: Add 4th Convolutional layer with relu activation with same padding
cnn_model.add(Conv2D(256, kernel_size=(3,3), padding='same',activation='relu'))
cnn_model.add(MaxPooling2D(pool_size=(2,2)))

# Add flatten layer
cnn_model.add(Flatten())

# Add FCNN
cnn_model.add(Dense(256,activation='relu'))
cnn_model.add(Dropout(0.5)) # Add dropout

# Add FCNN
cnn_model.add(Dense(24, activation='softmax'))

```

```python
# Model Summary
cnn_model.summary()
```

**Total trainable parameter is 13,239,896**

### CNN Model Compilation

```python
from tensorflow.keras.optimizers import Adam,RMSprop
from tensorflow.keras.losses import SparseCategoricalCrossentropy

cnn_model.compile(loss='categorical_crossentropy', optimizer=Adam(0.1),metrics=['accuracy'])
```

### CNN Model Training

```python
from tensorflow.keras.callbacks import ModelCheckpoint

check = ModelCheckpoint(filepath='cnn_model.weights.best.hdf5',verbose=1,save_best_only=True)
history = cnn_model.fit(X_train,y_train, epochs=50, validation_data=(X_test,y_test),callbacks=[check],verbose=2)
```

```python
cnn_model.evaluate(X_test,y_test)
```
